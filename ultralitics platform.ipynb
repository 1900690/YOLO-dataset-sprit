```python
#@title ②：AIの学習設定と実行
#@markdown 以下のフォームの数値を調整して、学習をスタートしてください。

#@markdown ---
#@markdown ### 🔑 接続とデータの設定
#@markdown Ultralytics HUBと連携するための秘密の鍵です。
APIキー = "ul_047bf01e4de56d2a373aaacad650bbef204f13b9" #@param {type:"string"}

#@markdown 使用するAIのベース（土台）となるモデルを指定します。
モデル名 = "ul://ultralytics/yolo26/yolo26n" #@param {type:"string"}

#@markdown 学習させたい画像データが置いてある場所を指定します。
データセット = "ul://flamingo-flamingo/datasets/eggplant-flower-fruit3" #@param {type:"string"}

#@markdown 完了した学習結果を保存するプロジェクト名です。
保存先プロジェクト = "flamingo-flamingo/eggplant-bbox" #@param {type:"string"}

#@markdown ---
#@markdown ### 🧠 学習の強さ（パラメータ）設定
#@markdown 学習を何周繰り返すか（多いほど賢くなりますが、時間はかかります）
繰り返し回数_エポック = 100 #@param {type:"integer"}

#@markdown 1回の学習で何枚の画像をまとめて処理するか（-1 はパソコンの性能に合わせて自動調整）
バッチサイズ = -1 #@param {type:"integer"}

#@markdown 読み込む画像のサイズ（基本は 640 のままでOKです）
画像サイズ = 640 #@param {type:"integer"}

import os
from ultralytics import YOLO

# 1. APIキーをシステムにセットする
os.environ["ULTRALYTICS_API_KEY"] = APIキー

# 2. 指定したモデル（AIの土台）を読み込む
print(f"🔄 モデル {モデル名} を読み込んでいます...")
model = YOLO(モデル名, task="detect")

# 3. フォームで設定した内容で学習をスタートする
print("🚀 学習をスタートします！")
model.train(
    data=データセット,
    epochs=繰り返し回数_エポック,
    batch=バッチサイズ,
    imgsz=画像サイズ,
    project=保存先プロジェクト,
)
```

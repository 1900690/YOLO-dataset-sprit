```python
#@title ②：AIの学習設定と実行
#@markdown ultralytics platformのLocal Trainningで指定される数値をコピペしてください

#@markdown ---
#@markdown ### 🔑 接続とデータの設定
#@markdown Ultralytics platformと連携するための秘密の鍵です。
APIキー = "" #@param {type:"string"}

#@markdown 使用するAIのベース（土台）となるモデルを指定します。
モデル名 = "" #@param {type:"string"}

#@markdown 学習させたい画像データが置いてある場所を指定します。
データセット = "" #@param {type:"string"}

#@markdown 完了した学習結果を保存するプロジェクト名です。
保存先プロジェクト = "" #@param {type:"string"}

#@markdown ---
#@markdown ### 🧠 学習の強さ（パラメータ）設定
#@markdown 学習を何周繰り返すか（多いほど賢くなりますが、時間はかかります）
繰り返し回数_エポック = None #@param {type:"integer"}

#@markdown 1回の学習で何枚の画像をまとめて処理するか（-1 はパソコンの性能に合わせて自動調整）
バッチサイズ = None #@param {type:"integer"}

#@markdown 読み込む画像のサイズ（基本は 640 のままでOKです）
画像サイズ = None #@param {type:"integer"}

import os
from ultralytics import YOLO

# 1. APIキーをシステムにセットする
os.environ["ULTRALYTICS_API_KEY"] = APIキー

# 2. 指定したモデル（AIの土台）を読み込む
print(f"🔄 モデル {モデル名} を読み込んでいます...")
model = YOLO(モデル名, task="detect")

# 3. フォームで設定した内容で学習をスタートする
print("🚀 学習をスタートします！")
model.train(
    data=データセット,
    epochs=繰り返し回数_エポック,
    batch=バッチサイズ,
    imgsz=画像サイズ,
    project=保存先プロジェクト,
)
```
